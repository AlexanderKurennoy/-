```python
#Завантаження бібліотеки
import numpy as np
```

```python
#Створіть одновимірний масив (вектор) із арифметичною прогресією, де кожне наступне число менше попереднього на 4,
#починаючи з 50, та виведіть його значення.
arr = np.arange(50, -1, -4)
print(arr)
```

```python
#Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його значеннями суми добутків кожного числа на його попереднє,
#починаючи з 1, та виведіть його значення.
matrix = np.arange(1, 10).reshape(3, 3) * (np.arange(1, 10).reshape(3, 3) - 1)
print(matrix)
```

```python
#Створіть масив розміром 3x3, заповніть його випадковими цілими числами в діапазоні від -5 до 5 та знайдіть
#мінімальний елемент.
matrix = np.random.randint(-5,6,size=(3,3))
min_n = np.min(matrix)
print(matrix)
print("Мінімальний елемент:", min_n)
```

```python
#Створіть масив розміром 3x3, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та обчисліть
#добуток елементів, що менше 0.2.
random_matrix = np.random.random((3, 3))
result = np.prod(random_matrix[random_matrix < 0.2])
print(random_matrix)
print(f"Добуток елементів менше 0.2: {result}")
```

```python
#Створіть два одновимірних масиви розміром 4, заповніть їх випадковими дійсними числами від 0 до 1
#та виконайте на них поелементні операції додавання, віднімання та множення.
array_a = np.random.rand(4)
array_b = np.random.rand(4)
plus = array_a + array_b
minus = array_a - array_b
dobutok = array_a * array_b
print("Array A:", array_a)
print("Array B:", array_b)
print("Сума:", plus)
print("Різниця:", minus)
print("Добуток:", dobutok)
```

```python
#Створіть дві матриці 3x3, заповніть їх довільними цілими числами від 5 до 10 та знайдіть їх добуток.
matrix_a = np.random.randint(5, 11, size=(3, 3))
matrix_b = np.random.randint(5, 11, size=(3, 3))
array = np.dot(matrix_a, matrix_b)
print("Матриця A:\n", matrix_a)
print("Матриця B:\n", matrix_b)
print("Array\n", array)
```

```python
#Створіть дві матриці розміром 2x3 та 3x3, заповніть їх випадковими цілими числами в діапазоні від
#1 до 5 та перемножте їх між собою.
matrix_a = np.random.randint(1, 6, size=(2, 3))
matrix_b = np.random.randint(1, 6, size=(3, 3))
array = np.dot(matrix_a, matrix_b)
print("Матриця A:\n", matrix_a)
print("Матриця B:\n", matrix_b)
print("Array:\n", array)
```

```python
#Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1
#до 8 та знайдіть її обернену матрицю.
matrix = np.random.randint(1, 9, size=(3, 3))
rotate_matrix = np.linalg.inv(matrix) if np.linalg.det(matrix) != 0 else None
print("Матриця:\n", matrix)
if rotate_matrix is not None:
    print("Обернена матриця:\n", rotate_matrix)
else:
    print("Матриця не обернена")
```

```python
#Створіть матрицю розміром 8x8, заповніть її випадковими дійсними
#числами в діапазоні від 50 до 100 та транспонуйте її.
matrix = np.random.uniform(50, 101, size=(8, 8))
tran_matrix = np.transpose(matrix)
print("Матриця:\n", matrix)
print("Транспонована матриця:\n", tran_matrix)
```

```python
#Створіть матрицю розміром 3x4 та вектор розміром 4x1, заповніть матрицю випадковими дійсними
#числами від -1 до 1, а вектор - випадковими цілими числами в діапазоні від -3 до 3 і знайдіть
#добуток матриці на вектор.
matrix = np.random.uniform(-1, 1, size=(3, 4))
vector = np.random.randint(-3, 4, size=(4, 1))
array = np.dot(matrix, vector)
print("Матриця:\n", matrix)
print("Вектор:\n", vector)
print("Добуток матриці на вектор:\n", array)
```

```python
#Створіть матрицю розміром 2x3 та вектор розміром 3x1, заповніть матрицю випадковими цілими числами
#в діапазоні від 1 до 10, а вектор - випадковими дійсними числами від 0 до 1, і знайдіть
#добуток матриці на вектор.
matrix = np.random.randint(1, 11, size=(2, 3))
vector = np.random.rand(3, 1)
array = np.dot(matrix, vector)
print("Матриця:\n", matrix)
print("Вектор:\n", vector)
print("Array:\n", array)
```

```python
#Створіть дві матриці розміром 5x5, заповніть їх випадковими цілими числами в
#діапазоні від 1 до 10 та знайдіть відношення всіх елементів першої матриці до відповідних
#елементів другої матриці (поелементне ділення).
matrix1 = np.random.randint(1, 11, size=(5, 5))
matrix2 = np.random.randint(1, 11, size=(5, 5))
dividing = np.divide(matrix1, matrix2)
print("Матриця A:\n", matrix1)
print("Матриця B:\n", matrix2)
print("Результат ділення:\n", dividing)
```

```python
#Створіть дві матриці розміром 4x4, заповніть їх випадковими цілими числами в діапазоні
#від -10 до 10 та знайдіть різницю їх елементів.
matrix1 = np.random.randint(-10, 11, size=(4, 4))
matrix2 = np.random.randint(-10, 11, size=(4, 4))
equation = np.subtract(matrix1, matrix2)
print("Матриця A:\n", matrix1)
print("Матриця B:\n", matrix2)
print("Різниця:\n", equation)
```

```python
#Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні
#від 1 до 10 та знайдіть суму елементів матриці, які менше за їх середнє
#арифметичне значення.
matrix = np.random.randint(1, 11, size=(3, 3))
mean = np.mean(matrix)
sum_below_mean = np.sum(matrix[matrix < mean])
print("Матриця:\n", matrix)
print(f"Середнє арифметичне: {mean}")
print(f"Сума елементів менше середнього: {sum_below_mean}")
```

```python
#Створіть матрицю розміром 3x3, заповніть її випадковими дійсними
#числами в діапазоні від -2 до 2 та знайдіть вектор-стовпчик, що містить
#середнє гармонічне значення елементів кожного рядка матриці.
matrix = np.random.uniform(-2, 2, size=(3, 3))
harmonic_mean = np.reciprocal(np.mean(np.reciprocal(matrix), axis=1)).reshape(-1, 1)
print("Матриця:\n", matrix)
print("Гармонічне середнє вектор:\n", harmonic_mean)
```
